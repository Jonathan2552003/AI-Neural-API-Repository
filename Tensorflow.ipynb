```python
import tensorflow as tf
import numpy as np
```

```python
celsius = np.array([40,-10,0,8,15,22,38], dtype=float)
fareheit = np.array([104,14,32,46.4,59,71.6,100.4], dtype=float)
```

```python
capa = tf.keras.layers.Dense(units = 1, input_shape = [1])
modelo = tf.keras.Sequential([capa])
```

```python
modelo.compile(optimizer= tf.keras.optimizers.Adam(0.01), loss = "mean_squared_error")
```

```python
#Ahora si vamos a entrenarlo
print("Comenzamos entrenamiento")

historial = modelo.fit(celsius,fareheit, epochs=1000, verbose =False)
print("Modelo entrenado")

```

```python
#Aqui vamos a ahcer unos resultados de entrenamiento de la red en cada vuelta que dio
#Mas bien, una grafica de su aprendizaje

import matplotlib.pyplot as plt
plt.xlabel("Entrenamiento")
plt.ylabel("Magnitud de perdida")
plt.plot(historial.history["loss"])
```

```python
print("Aquí vamos a ver si quedó bien entrenada")

# Cambia 'predic' a 'predict' en la siguiente línea
resultado = modelo.predict([100])

print("El resultado es"+str(resultado) +"Fahrenheit")
```
